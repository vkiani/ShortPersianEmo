```
import re
def preprocess_text(text):
  
  text = re.sub(r"(?:@\S*|http(?=.*://)\S*)", "", text) #Remove Hashtag and username
  text = re.sub(r'http\S+', '', text) #Remove link
  text = text.replace("'",' ')
  text = text.replace('"',' ')
  text = re.sub('\s+', ' ', text)   # remove whitespaces

  return text
```
