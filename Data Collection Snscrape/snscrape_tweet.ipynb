```
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
```

```
import re
def preprocess_text(text):
  
  text = re.sub(r"(?:@\S*|http(?=.*://)\S*)", "", text) #Remove Hashtag and username
  text = re.sub(r'http\S+', '', text) #Remove link
  text = text.replace("'",' ')
  text = text.replace('"',' ')
  text = re.sub('\s+', ' ', text)   # remove whitespaces

  return text
```

```
maxTweets = 300
tweets_list = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('داعش تجاوز (-😊 -😂 -😁 -😆 -😅 -😃) lang:fa').get_items()):  # 

  if i > maxTweets or len(tweets_list) > 1000:
    break
  if len(tweets_list) > 0 and len(tweets_list) % 1000 == 0:
    print(len(tweets_list))
  if len(preprocess_text(tweet.rawContent)) < 100 and len(preprocess_text(tweet.rawContent)) > 30:
      tweets_list.append(preprocess_text(tweet.rawContent))
```
